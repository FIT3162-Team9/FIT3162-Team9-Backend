## Filling the missing humidity9am data

Since there are some missing date in our data so we need to fill the missing dates and other values will be NaN

```python
temp_weather_data = pd.DataFrame()
for each_location in location:
    weather_location_data = weather_data.loc[weather_data['Location']==each_location]
    range_of_date=idx=pd.date_range(pd.to_datetime(start_date.loc[start_date['Location']==each_location,'Date'].values[0]),pd.to_datetime(end_date.loc[end_date['Location']==each_location,'Date'].values[0]))
    weather_location_data =weather_location_data.set_index('Date').reindex(idx).reset_index().rename(columns={'index':'Date'})
    weather_location_data.Location=weather_location_data.Location.fillna(each_location)
    
    temp_weather_data=pd.concat([temp_weather_data,weather_location_data])
```

```python
print("After filling the missing date we have total rows of",len(temp_weather_data))
```

```python
temp_weather_data=temp_weather_data.reset_index().drop('index',1)
```

### Take a look at the missing humidity value

First find all the Nan values in Humidity9am col

```python
missing_humidity9am_data = temp_weather_data[np.isnan(temp_weather_data['Humidity9am'])][['Date','Location']]
```

### Filling missing data

```python
#### from pandas.tseries.offsets import DateOffset
humidity_mean=temp_weather_data.describe().loc['mean','Humidity9am']
for index, row in missing_humidity9am_data.iterrows():
    date,location = row
    #if the year is the start year, we look up humidity data in the next years 2008, 2009,... (until we find not-null value)
    if date.strftime('%Y')==pd.to_datetime(start_date.loc[start_date['Location']==location,'Date'].values[0]).strftime('%Y'):
        date_offset=date + DateOffset(years=1)
        find_notnull=False
        while date_offset<=pd.to_datetime(end_date.loc[end_date['Location']==location,'Date'].values[0]):
            nextyear_humidity9am = temp_weather_data.loc[(temp_weather_data['Date']==date_offset) & (temp_weather_data['Location']== location),'Humidity9am'].values[0]
            if np.isnan(nextyear_humidity9am)==False :
                temp_weather_data.at[index,'Humidity9am'] = nextyear_humidity9am
                find_notnull=True
                break
            date_offset=date_offset + DateOffset(years=1)
        #if we cannot find not null value
        if find_notnull==False:
            temp_weather_data.at[index,'Humidity9am'] = humidity_mean
            
    #if the year is upper limit year, we look up humidity data in the previous years 2016, 2015,...(until we find not-null value)
    elif date.strftime('%Y')==pd.to_datetime(end_date.loc[end_date['Location']==location,'Date'].values[0]).strftime('%Y'):
        date_offset=date - DateOffset(years=1)
        find_notnull=False
        while date_offset>=pd.to_datetime(start_date.loc[start_date['Location']==location,'Date'].values[0]):
            lastyear_humidity9am =  temp_weather_data.loc[(temp_weather_data['Date']==date_offset) & (temp_weather_data['Location']== location),'Humidity9am'].values[0]
            if np.isnan(lastyear_humidity9am)==False:
                temp_weather_data.at[index,'Humidity9am'] = lastyear_humidity9am
                find_notnull=True
                break
            date_offset=date_offset - DateOffset(years=1)
        #if we cannot find not null value
        if find_notnull==False:
            temp_weather_data.at[index,'Humidity9am'] = humidity_mean

    else:
        filling_missing=0
        date_offset=date + DateOffset(years=1)
        nextyear_humidity9am=-1
        while date_offset<=pd.to_datetime(end_date.loc[end_date['Location']==location,'Date'].values[0]):
            nextyear_humidity9am = temp_weather_data.loc[(temp_weather_data['Date']==date_offset) & (temp_weather_data['Location']== location),'Humidity9am'].values[0]
    
            if np.isnan(nextyear_humidity9am)==False:
                break
            date_offset=date_offset + DateOffset(years=1)
        date_offset= date-DateOffset(years=1)
        lastyear_humidity9am=-1
        while date_offset>=pd.to_datetime(start_date.loc[start_date['Location']==location,'Date'].values[0]):
            lastyear_humidity9am =  temp_weather_data.loc[(temp_weather_data['Date']==date_offset) & (temp_weather_data['Location']== location),'Humidity9am'].values[0]
            if np.isnan(lastyear_humidity9am)==False:
                break
            date_offset=date_offset - DateOffset(years=1)
            
        if np.isnan(nextyear_humidity9am) == True and np.isnan(lastyear_humidity9am)==True:
            filling_missing = humidity_mean
        elif np.isnan(nextyear_humidity9am) == True and np.isnan(lastyear_humidity9am)==False:
            filling_missing = lastyear_humidity9am
        elif np.isnan(nextyear_humidity9am) ==False and np.isnan(lastyear_humidity9am)==True:
            filling_missing = nextyear_humidity9am
        else:
            filling_missing = (nextyear_humidity9am + lastyear_humidity9am)/2
        temp_weather_data.loc[(temp_weather_data['Date']==date) & (temp_weather_data['Location']== location),'Humidity9am'] = filling_missing
        
    
```
